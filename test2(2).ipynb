```python
import cv2

# Load the image and Haar cascade file
img = cv2.imread('f49fd48d6dbfd2aa202af842df1966b2.jpg')
if img is None:
    raise FileNotFoundError("The image file was not found or could not be opened.")

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
if face_cascade.empty():
    raise FileNotFoundError("The Haar cascade XML file was not found or could not be loaded.")

# Define the face detection function
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.3, minNeighbors=3)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (0, 0, 255), 2)
        
    return face_img

# Perform face detection
face_img = detect_face(img)

# Save the images to files
cv2.imwrite('original_image.jpg', img)
cv2.imwrite('detected_face.jpg', face_img)
```
